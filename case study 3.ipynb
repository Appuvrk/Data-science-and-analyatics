```python
import numpy as np 
import pandas as pd
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
import scipy.stats as stats

```

```python
data = pd.read_csv('Downloads/Sales_add.csv')
data
```

```python
data.isnull().sum()
```

```python
data1 = data.rename(columns ={'Sales_before_digital_add(in $)': 'Sales Before', 'Sales_After_digital_add(in $)': 'Sales After'})
data1
```

# 1. INCREASE IN SALES AFTER STEPPING INTO DIGITAL MARKETING

```python
# H0 = No increase in sales after stepping into digital marketing
# H1 = Increase in sales after stepping into digital marketing 
alpha = 0.05
t_value,p_value = stats.ttest_rel(data1['Sales After'], data1['Sales Before'])
print("The calculated t score & p-value are : \n \nt-value = %0.3f \np-value = %0.3f \n" % (t_value,p_value))
```

```python
if (p_value > alpha):
    print("Fails to reject the null hypoyhesis")
else:
    print("Rejecting the null hypothesis")
```

# 2. DEPENDENCY BETWEEN REGION AND MANAGER2.

```python
# H0 = No relationship between regionn and manager
# H1 = There is a relationship between region and manager

```

```python
data2 = pd.crosstab(data1['Region'],data['Manager'])
data2
```

```python
observed_values = data2.values
print('observed values:-\n',observed_values)
```

```python
value = stats.chi2_contingency(data2)
value
```

```python
expected_value = value[3]
expected_value
```

```python
no_of_rows = len(data2.iloc[0:3,0])
no_of_columns = len(data2.iloc[0,0:3])
ddof = (no_of_rows-1)*(no_of_columns-1)
print('Degrees of Freedom:',ddof)
alpha = 0.05
from scipy.stats import chi2
```

```python
chi_square = sum([(o-e)**2./e for o,e in zip(observed_values,expected_value)])
chi_square_statistics = chi_square[0] + chi_square[1]
print('chi-Square statistics:',chi_square_statistics)
critical_value = chi2.ppf(q = 1-alpha,df = ddof)
print('Critical_value:',critical_value)
```

```python
p_value = 1-chi2.cdf(x = chi_square_statistics,df = ddof)
print('p_value:',p_value)
print('Significance Level:',alpha)
print('Degrees of Freedom:',ddof)
```

```python
if (p_value > alpha):
    print('Fails to reject the null hypothesis')
else:
    print('Rejecting the null hypothesis')
```
